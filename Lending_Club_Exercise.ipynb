### We have completed removing outliner for all critical columns, now we will start work on creating binning for loan amount, interest rate, annual income and employment lenngth for apply EDA process and creating new columns 

```python
def ln_amt(x):
    if x <= 5500:
        return "Small"
    elif 5500 < x <= 10000:
        return "Medium"
    elif 10000 < x <= 15000:
        return "Large"
    else:
        return "Very Large"
    
#lending.loan_amnt.apply(ln_amt).value_counts()
lending['loan_type']=lending.loan_amnt.apply(ln_amt)
```

```python
lending.loan_type.value_counts()
```

### To simplify bucketting converting float into integer for interest rate which will make easy

```python
lending['int_rate']=lending['int_rate'].astype('int64')
```

```python
lending.head(10)
```

### We will do binning for interest rate

```python
def interest(x):
    if x <= 10:
        return "Low"
    elif 10 < x <= 12:
        return "Medium"
    elif 12 < x <= 15:
        return "High"
    else:
        return "Very High"
#lending.int_rate.apply(interest).value_counts()
lending['int_range']=lending.int_rate.apply(interest)
```

```python
lending.dtypes
```

### Let's do binning for annual income

```python
def aincome(x):
    if x <= 40000:
        return "Low"
    elif 40000 < x <= 60000:
        return "Medium"
    elif 60000 < x <= 90000:
        return "High"
    else:
        return "Very High"
lending['income_range']=lending.annual_inc.apply(aincome)
```
